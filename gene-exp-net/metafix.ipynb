```python
import pandas as pd
import numpy as np

orig_meta = pd.read_csv('Arabidopsis_metadata.tsv', sep='\t', header=0, index_col=0)
metafixr=pd.read_csv('metafix.csv', header=0) #note that i changed '0' to 'index' in the csv to make this work
metafixr.head()
```
